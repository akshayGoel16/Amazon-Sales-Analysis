```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Amazon Sale Report.csv')
```

```python
df.describe()
```

```python
# DATA CLEANING
```

```python
df.drop(['New', 'PendingS'], axis=1, inplace=True)
df
```

```python
df.describe()
```

```python
pd.isnull(df)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
df
```

```python
df.columns
```

```python
df['ship-postal-code'] = df['ship-postal-code'].astype('int')
```

```python
df['ship-postal-code'].dtype
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.rename(columns = {'Qty':'Quantity'}, inplace=True)
df.columns
```

```python
df.describe()
```

```python
df.describe(include='object')
```

```python
df['Quantity'].describe()
```

```python
# DATA ANALYSIS
```

```python
ax=sns.countplot(x='Size', data=df)

for bars in ax.containers:
    ax.bar_label(bars)

plt.title('Apparel Chart')
plt.show()
```

```python
#From the above Graph, we can cleary see that 'M' size has the maximum count.
```

```python
S_Qty = df.groupby(['Size'], as_index=False)['Quantity'].sum().sort_values(by='Quantity', ascending=True)

ax=sns.barplot(x='Size', y='Quantity', data=S_Qty, color='green')

for bars in ax.containers:
    ax.bar_label(bars)

plt.title('Apparel Chart')
plt.show()
```

```python
#From the above Graph, we can cleary see that 'M' size has the maximum sale.
df.columns
```

```python
sns.countplot(data = df, x='Courier Status', hue='Status')
plt.xlabel('Status')
plt.ylabel('Count')
plt.title('Status Chart')
```

```python
df['Size'].hist()
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Apparel Chart')
plt.show()
```

```python
df['Category'] =df['Category'].astype(str)
column_data = df['Category']
plt.figure(figsize=(10,5))
plt.hist(column_data, bins=10, edgecolor="Black")
plt.xticks(rotation=90)
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Apparel Chart')
plt.show()
```

```python
# Most Sold Category is T-shirt
```

```python
B2B_Check = df['B2B'].value_counts()
```

```python
plt.pie(B2B_Check, labels = B2B_Check.index, autopct = '%1.1f%%')
plt.title('Apparel Chart')
plt.show()
```

```python
x_data = df['Category']
y_data = df['Size']
```

```python
plt.scatter(x_data, y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Apparel Chart')
plt.show()
```

```python
plt.figure(figsize=(12,6))
top_State = df['ship-state'].value_counts().head(7)
sns.countplot(data=df[df['ship-state'].isin(top_State.index)], x='ship-state')
plt.title('State Chart')
plt.xticks(rotation=90)
plt.show()
```

```python
# MOST_SALES comes from the state of MAHARASHTRA
```

```python
# CONCLUSION:-
# The State of Maharashtra has a large Customer Base.
# T-shirt & Shirt are the two categories that are available in all the sizes.
# Majority of business are not running on a B2B model except for a negligible number.
# The Most Sold apparel in the categories is T-shirt.
# Majority of the orders are successfully delivered to their buyers.
# Among different Size Categories, 'M' size has the highest Demand and Sale.        
```
